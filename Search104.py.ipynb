```python
from collections import Counter
from bs4 import BeautifulSoup
import requests

cnt = Counter()

for i in range(1,2):
    res = requests.get('https://www.104.com.tw/jobbank/joblist/joblist.cfm?keyword=%E8%BB%9F%E9%AB%94&jobsource=n104bank1&ro=0&order=1&page='+str(i))
    soup = BeautifulSoup(res.text, 'lxml')
    for hr in soup.select(".job_name > a"):
        cp = hr['href'].split("jobno=")[1].split("&jobsource")[0]
        res2 = requests.get('https://www.104.com.tw/job/?jobno='+cp+'&jobsource=104_hotorder')
        soup=BeautifulSoup(res2.text,'lxml')
        for tr in soup.select('.tool > a'):           
            if tr.text in cnt:
                cnt[tr.text]+=1
            else:
                cnt[tr.text]=1
print(cnt.most_common(20))
```
