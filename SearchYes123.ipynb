```python
def strdict(data_str):
    data = {}
    for row in data_str.split('\n'):
        kv_list = row.split(":")
        data[kv_list[0]] = kv_list[1]
    return data
```

```python
import requests
from bs4 import BeautifulSoup
from collections import Counter

cnt = Counter()

url = "https://www.yes123.com.tw/admin/job_refer_list.asp"
for j in range(0, 300, 20):
    form_data = strdict(data_str(j))
    res = requests.post(url , data=form_data)
    res.encoding = 'utf8'
    soup = BeautifulSoup(res.text , 'lxml')
    #print(soup.select('.jobname')[0]['href'])
    for a in soup.select('.jobname'):
        cp = a['href']
        uri = "https://www.yes123.com.tw/admin/"
        URL = uri + cp
        res2 = requests.post(URL , data=form_data)
        soup = BeautifulSoup(res2.text , 'lxml')
        for tr in soup.select('.comp_detail > ul > li'):
            #print(tr.text)
            if tr.text[:5] == '電腦技能：':
            #print(tr.text[6:].replace('公室應用：',' ').replace('作業系統：',' ').replace('／',' ').replace('未填寫',' ').split())
                for tr2 in tr.text[6:].replace('公室應用：',' ').replace('作業系統：',' ').replace('／',' ').replace('未填寫',' ').replace('式設計：',' ').replace('輕型機車、普通重型機車、普通小型車',' ').split():
                    if tr2 in cnt:
                        cnt[tr2]+=1
                    else:
                        cnt[tr2]=1
            else:
                pass
print(cnt.most_common(20))
```
